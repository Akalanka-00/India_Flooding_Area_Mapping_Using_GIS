```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
```

```python
def read_csv(file_path: str):

    try:
        df = pd.read_csv(file_path)
        return df  # Return the DataFrame
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return None
    except Exception as e:
        print(f"Error: {e}")
        return None
```

```python
csv_file_path = './flood_risk_dataset_india_new.csv'  # Replace with the actual path to your CSV
```

```python
df_new = read_csv(csv_file_path)
```

```python
print(df_new)
```

```python
geo_df = geopandas.GeoDataFrame(df_new, geometry=geopandas.points_from_xy(df_new.Longitude, df_new.Latitude))
```

```python
url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"
world = geopandas.read_file(url)
```
